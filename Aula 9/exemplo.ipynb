```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1)
nivel = np.array([50,60,70,45,80,85]).reshape(-1,1)
```

```python
modelo = LinearRegression()
modelo.fit(vazao, nivel)
```

```python
#vazao_teste = np.array([15,25,33]).reshape(-1,1)
vazao_teste = np.array([50,60,10,120,135,155]).reshape(-1,1)
```

```python

import matplotlib.pyplot as plt
nivel_previsto = modelo.predict(vazao_teste)
```

```python
for i in range(len(vazao_teste)):
    print(f"Vazao: {vazao_teste[i][0]} m3/mim - nivel previsto: {nivel_previsto[i]} metros")

vazao_teste = [x[0] for x in vazao_teste]  # Extrair os valores de vazao_teste para uma lista separada

plt.scatter(vazao, nivel, color='blue', label='dados de treinamento')
plt.plot(vazao_teste, nivel_previsto, color='red', label='dados de teste')  # Usar nivel_previsto em vez de nivel
plt.xlabel("vazao de agua (m3/mim)")
plt.ylabel("nivel do tanque (metros)")
plt.title("relação entre vazao de agua e nivel do tanque")
plt.legend()
plt.grid(True)
plt.show()
```
