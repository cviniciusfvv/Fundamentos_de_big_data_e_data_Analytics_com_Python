```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
calor = np.array([100,200,300,400,500,600,700]).reshape(-1,1)
temperatura = np.array([30,35,40,45,60,100,150]).reshape(-1,1)
```

```python
modelo = LinearRegression()
modelo.fit(calor, temperatura)
```

```python
#vazao_teste = np.array([15,25,33]).reshape(-1,1)
carlor_teste = np.array([50,60,10,120,135,155]).reshape(-1,1)
```

```python

import matplotlib.pyplot as plt
temperatura_prevista = modelo.predict(carlor_teste)
```

```python
for i in range(len(carlor_teste)):
    print(f"Vazao: {carlor_teste[i][0]} Temperatuda - nivel previsto: {temperatura_prevista[i]} Cª")

carlor_teste = [x[0] for x in carlor_teste]  # Extrair os valores de vazao_teste para uma lista separada

plt.scatter(calor, temperatura, color='blue', label='dados de treinamento')
plt.plot(carlor_teste, temperatura_prevista, color='red', label='dados de teste')  # Usar nivel_previsto em vez de nivel
plt.xlabel("calor emitido Temperatuda")
plt.ylabel("temperatura (Cª)")
plt.title("relação entre calor e temperatura")
plt.legend()
plt.grid(True)
plt.show()
```
