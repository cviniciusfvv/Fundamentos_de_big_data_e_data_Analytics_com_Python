Aplicar o algoritimo de KMeans para agrupar os dias de maior movimentação e plotar graficos de disperção
dia da semana = [1,2,3,4,5,6,7]
numero de clientes = [15,20,30,100,350,500,700]

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dias_semana = np.array([1, 2, 3, 4, 5, 6, 7])
num_clientes = np.array([15, 20, 30, 100, 350, 500, 700])

dados = np.column_stack((dias_semana, num_clientes))
```

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(dados)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(dias_semana, num_clientes, c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Clientes')
plt.title('Agrupamento dos Dias de Maior Movimentação')
plt.show()
```

```python
dados_df = pd.DataFrame(dados, columns=['Dia da Semana', 'Número de Clientes'])
dados_df['Grupo'] = labels
grouped_data = dados_df.groupby('Grupo').mean()
print(grouped_data)
```
