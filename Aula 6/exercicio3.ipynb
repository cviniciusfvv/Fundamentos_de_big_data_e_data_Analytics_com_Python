```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df=pd.read_csv('monitoramento_tempo.csv',sep=',')
```

```python
df.info()
```

```python
df['data']=pd.to_datetime(df['data'])
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
#eixo.set_ylim(200,500)
eixo.plot(df['data'],df['pressão'],color= 'g')
eixo.grid(True)
eixo.set_title('pressão no momento',fontsize=25, pad =20)
eixo.set_ylabel('pressão', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['pressão'],loc='best')
eixo.axhline(max(df['pressão']),color='k',linestyle= '--')
eixo.axhline(min(df['pressão']),color='k',linestyle= '--')
x1=df['data'][df['pressão'].idxmax()]
y1 = max(df['pressão'])
x2= x1=df['data'][df['pressão'].idxmax()]
y2= max(df['pressão']) -5
eixo.annotate('Máximo',xy=(x1,y1),fontsize=20,xytext=(x2,y2), arrowprops=dict(facecolor='k'))
x1=df['data'][df['pressão'].idxmin()]
y1 = min(df['pressão'])
x2= x1=df['data'][df['pressão'].idxmin()]
y2= min(df['pressão']) +5
eixo.annotate('Mínimo',xy=(x1,y1),fontsize=20,xytext=(x2,y2), arrowprops=dict(facecolor='k'))
```
