2) Você é um cientista de dados em um laboratório médico e recebeu
um conjunto de dados contendo os resultados de exames de sangue
de hemograma de diferentes pacientes. Seu objetivo é utilizar o
algoritmo DBSCAN para identificar grupos de pacientes com perfis de
hemograma semelhantes.
Lecócitos = 2000,4000,5000,6500,
Plaquetas = 100000,20000,80000,145000,
Linfócitos = 2.3,4.5,6,5

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
```

```python
# Dados dos pacientes
leucocitos = [2000, 4000, 5000, 6500]
plaquetas = [100000, 20000, 80000, 145000]
linfocitos = [2.3, 4.5, 6, 5]
```

```python
# Criar o DataFrame com os resultados do hemograma
df = pd.DataFrame({'Leucócitos': leucocitos, 'Plaquetas': plaquetas, 'Linfócitos': linfocitos})

# Aplicar o algoritmo DBSCAN
modelo = DBSCAN(eps=1000, min_samples=2)
predicoes = modelo.fit_predict(df)

# Adicionar a coluna de cluster ao DataFrame
df['Cluster'] = predicoes

# Criar a tabela 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotar os pontos na tabela 3D com cores diferentes para cada cluster
for cluster in df['Cluster'].unique():
    dados_cluster = df[df['Cluster'] == cluster]
    ax.scatter(dados_cluster['Leucócitos'], dados_cluster['Plaquetas'], dados_cluster['Linfócitos'], label=f'Cluster {cluster}')

# Configurar os rótulos dos eixos
ax.set_xlabel('Leucócitos')
ax.set_ylabel('Plaquetas')
ax.set_zlabel('Linfócitos')

# Adicionar uma legenda
ax.legend()

# Exibir a tabela 3D
plt.show()
```
